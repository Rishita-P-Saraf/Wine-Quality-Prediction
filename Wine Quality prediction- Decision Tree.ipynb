```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Read the Dataset

```python
df = pd.read_csv('WineQT.csv')
df.head()
```

```python
df.columns
```

```python
df["quality"].unique()
```

**Note: EDA is skipped in this part of the implementation**

## Splitting the data into X(independent features) and Y(target column)

```python
X= df.drop(columns="quality", axis=1)
Y= df["quality"]
```

```python
X
```

```python
Y
```

## Split data into training and testing dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test= train_test_split(X,Y, test_size=0.30, random_state=42)
```

```python
X_train
```

# Data Modeling

```python
from sklearn.tree import DecisionTreeClassifier
model= DecisionTreeClassifier()
model.fit(X_train,Y_train)
```

# Model Predictions

```python
Y_pred= model.predict(X_test)
```

```python
Y_pred
```

```python
from sklearn import tree
plt.figure(figsize=(20, 20))
tree.plot_tree(model, filled=True)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(X_train, Y_train)
```

```python
rf.score(X_test, Y_test)
```

# Hyperparameter Tuning

- GridSearchCV
- RandomizedSearchCV

```python
grid_param= {
    "n_estimators": [50,100,120],
    "criterion":["gini", "entropy"],
    "max_depth": range(5),
}
```

```python
from sklearn.model_selection import GridSearchCV
grid_search_rf= GridSearchCV(param_grid= grid_param, cv=10, n_jobs=6, verbose=1, estimator=rf)
grid_search_rf.fit(X_train, Y_train)
```

```python
grid_search_rf.best_params_
```

```python
rf_new= RandomForestClassifier(n_estimators= 120, criterion= 'gini', max_depth= 4)
rf_new.fit(X_train, Y_train)
```

```python
rf_new.score(X_test, Y_test)
```
