## Splitting the data into X(input features) and Y(target column)

```python
X=df.drop("quality", axis=1)
Y= df["quality"]
```

```python
X
```

```python
Y
```

## Standardization

#### Outlier Detection -> Boxplot

```python
fig, ax= plt.subplots(figsize=(15,15))
sns.boxplot(data=X, ax=ax)
```

## Split data into training and testing dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test =train_test_split(X,Y,test_size=0.2, random_state=44)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train= scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
fig, ax= plt.subplots(figsize=(15,15))
sns.boxplot(data=X_train, ax=ax)
```

# Data Modeling -SVC Classifier

```python
from sklearn.svm import SVC
svm_classifier=SVC()
svm_classifier.fit(X_train, Y_train)
svm_classifier.predict(X_test)
```

```python
svm_classifier.score(X_test, Y_test)
```

## Hyperparameter Tuning 
-GridSearchCV or RandomizedSearchCV

```python
param_grid={
    'C':[0.1,1,10],
    'kernel': ['linear', 'rbf', 'poly'],
    'gamma':['scale', 'auto'],
    'degree':[2,3,4]
}

## to apply gridSearchCV
from sklearn.model_selection import GridSearchCV
grid_search=GridSearchCV(svm_classifier, param_grid, cv=5)
grid_search.fit(X_train,Y_train)
```

```python
grid_search.best_params_  # Which combo was best?
```

```python
grid_search.best_score_   # What was the best score?
```

```python
## to apply RandomizedSearchCV
from sklearn.model_selection import RandomizedSearchCV
random_search=RandomizedSearchCV(svm_classifier, param_grid, cv=5)
random_search.fit(X_train,Y_train)
```

```python
random_search.best_params_
```

```python
## Train model with best optimal parameters
svm_classifier_optimized_params=SVC(C=1, degree=2, gamma='scale', kernel='rbf')
svm_classifier_optimized_params.fit(X_train,Y_train)
```

```python
svm_classifier_optimized_params = grid_search.best_estimator_
svm_classifier_optimized_params.score(X_test, Y_test)
```
